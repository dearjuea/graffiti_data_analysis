```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
graf_df = pd.read_csv("Parks_Graffiti_Report_-_Dataset.csv")
graf_df = graf_df.rename(columns = {'Location Name':"location_name","Work Performed Date":"work_date","Location Address":"location_address","Issue":"issue",
                                    "Description":"descr","Labor Cost":"labor","Material Cost":"material","Total Cost":"total_cost","Location":"location"})
graf_df["work_date_y_m"] = pd.to_datetime(graf_df.work_date).dt.strftime('20%y-%m')
#by_location = graf_df.groupby("location_name").describe().iloc[:,[0,1]]
#by_location['loc_freq_group'] = pd.cut(x= by_location.iloc[:,0], bins = [0.0,5.0,20.0,60.0],labels = ['1-5','5-20','20-60'])
#plt.hist(by_location['loc_freq_group'], bins = 3)
#plt.title('frequency of incidents by location')

by_time = graf_df.groupby("work_date_y_m").describe()


plt.plot(by_time.index,by_time.iloc[:,0])
plt.ylim(0,50)
plt.title('trend of incidents')
plt.xticks(rotation = 90,fontsize = 7)
plt.show()
```

```python
by_time.head()
cost_by_time = graf_df['total_cost'].groupby(graf_df.work_date_y_m).sum()
plt.plot(cost_by_time.index, cost_by_time)
plt.title('trend of total cost')
plt.xticks(rotation = 90,fontsize = 7)
plt.show()
```

```python
fig, ax = plt.subplots()

ax.bar(graf_df.index,graf_df.labor,label='labor')
ax.bar(graf_df.index,graf_df.material,label='material')

ax.set_ylabel('labor cost vs. material cost')
ax.set_title('Cost breakdown by incidents')
ax.legend()

plt.show()
```
